```python
from dataIngestDemo import KaggleDataGetter
```

Define Constants

```python
CENSUS_DATASET_NAME = "muonneutrino/us-census-demographic-data"
CENSUS_FILE_NAME = "acs2017_county_data.csv"

COVID_DATASET_NAME = "nightranger77/covid19-state-data"
COVID_FILE_NAME = "COVID19_state.csv"
```

Load datasets

```python
dataGetter = KaggleDataGetter()
censusDf = dataGetter.getAsDataframe(CENSUS_DATASET_NAME, CENSUS_FILE_NAME)
covidDf = dataGetter.getAsDataframe(COVID_DATASET_NAME, COVID_FILE_NAME)
```

Aggregate census data by state

```python
censusPopByStateDf = censusDf[["State", "TotalPop"]].groupby(["State"]).sum().reset_index()
censusPopByStateDf
```

Compare Population Values
NOTE: PR is dropped because it is not in the census dataset

```python
populationDf = covidDf[["State", "Population"]] \
    .merge(censusPopByStateDf, on="State") \
    .rename(columns={"Population": "pop_covid_data", "TotalPop": "pop_census"})

populationDf["pop_diff"] = populationDf["pop_covid_data"] - populationDf["pop_census"]
populationDf["pop_diff_pct"] = (populationDf["pop_diff"]/populationDf["pop_census"]).round(3)

populationDf
```

Mean population diff percent

```python
populationDf["pop_diff_pct"].mean()
```

```python
populationDf["pop_diff_pct"].median()
```
